```python
import numpy as np
import matplotlib.pyplot as plt

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k
x_k = np.log(p_k / (1 - p_k))

# Définir la fonction de densité de probabilité fx pour λ = 1
def fx(x, λ=1):
    return (λ / 4) * (1 / np.cosh(λ * x / 2)**2)

# Valeurs pour x pour tracer la fonction de densité
x_values = np.linspace(-5, 5, 1000)
pdf_values = fx(x_values)

# Dessiner l'histogramme de la densité de l'échantillon {x_k}
plt.figure(figsize=(10, 6))
plt.hist(x_k, bins=100, density=True, alpha=0.5, color='blue', label='Histogramme des x_k')
plt.plot(x_values, pdf_values, color='red', label='Densité théorique f_x(x; λ=1)')
plt.title('Histogramme de la densité des valeurs x_k avec superposition de f_x(x; λ=1)')
plt.xlabel('x_k')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k
x_k = np.log(p_k / (1 - p_k))

# Définir la fonction de densité de probabilité fx
def fx(x, λ):
    return (λ / 4) * (1 / np.cosh(λ * x / 2)**2)

# Définir la fonction de log-vraisemblance négative
def negative_log_likelihood(λ, x_k):
    if λ <= 0:  # λ must be positive
        return np.inf
    likelihood = -np.sum(np.log(fx(x_k, λ)))
    return likelihood

# Estimer λ en minimisant la log-vraisemblance négative
initial_guess = 1.0
result = minimize(negative_log_likelihood, initial_guess, args=(x_k,), bounds=[(0.001, None)])

# Récupérer la valeur estimée de λ
lambda_hat = result.x[0]

# Afficher le résultat
print(f"Estimation du paramètre λ : {lambda_hat}")

# Dessiner l'histogramme avec la densité estimée
x_values = np.linspace(-5, 5, 1000)
pdf_values_estimated = fx(x_values, lambda_hat)

plt.figure(figsize=(10, 6))
plt.hist(x_k, bins=100, density=True, alpha=0.5, color='blue', label='Histogramme des x_k')
plt.plot(x_values, pdf_values_estimated, color='red', label=f'Densité estimée f_x(x; λ={lambda_hat:.2f})')
plt.title('Histogramme de la densité des valeurs x_k avec superposition de f_x(x; λ estimé)')
plt.xlabel('x_k')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k
x_k = np.log(p_k / (1 - p_k))

# Définir la fonction de densité de probabilité fx
def fx(x, λ):
    return (λ / 4) * (1 / np.cosh(λ * x / 2)**2)

# Définir la fonction de log-vraisemblance négative
def negative_log_likelihood(λ, x_k):
    if λ <= 0:  # λ must be positive
        return np.inf
    likelihood = -np.sum(np.log(fx(x_k, λ)))
    return likelihood

# Estimer λ en minimisant la log-vraisemblance négative
initial_guess = 1.0
result = minimize(negative_log_likelihood, initial_guess, args=(x_k,), bounds=[(0.001, None)])
lambda_hat = result.x[0]

# Maximum log-likelihood
max_log_likelihood = -result.fun

# Évaluer la log-vraisemblance pour une plage de λ
lambda_values = np.linspace(0.1, 5, 100)
log_likelihoods = -np.array([negative_log_likelihood(λ, x_k) for λ in lambda_values])

# Trouver les valeurs de λ pour ∆lnL = 1/2
delta_lnL = max_log_likelihood - 0.5
lambda_range = lambda_values[(log_likelihoods >= delta_lnL)]

# Calculer l'incertitude σˆλ
if len(lambda_range) > 0:
    lambda_lower = lambda_range[0]
    lambda_upper = lambda_range[-1]
    sigma_lambda_hat = (lambda_upper - lambda_lower) / 2
else:
    sigma_lambda_hat = 0  # No valid range found

# Afficher les résultats
print(f"Estimation du paramètre λ : {lambda_hat}")
print(f"Incertitude estimée σˆλ : {sigma_lambda_hat}")

# Vérifier la cohérence avec λ = 1
expected_lambda = 1
consistent = (lambda_hat - sigma_lambda_hat <= expected_lambda <= lambda_hat + sigma_lambda_hat)
print(f"Consistance avec λ = 1 : {'Oui' if consistent else 'Non'}")

# Optionnel : tracer la log-vraisemblance
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, log_likelihoods, label='Log-vraisemblance')
plt.axhline(delta_lnL, color='red', linestyle='--', label='ΔlnL = 1/2')
plt.axvline(lambda_hat, color='green', linestyle='--', label=f'λ̂ = {lambda_hat:.2f}')
plt.axvline(lambda_lower, color='orange', linestyle='--', label=f'λ_{lower:.2f}')
plt.axvline(lambda_upper, color='orange', linestyle='--', label=f'λ_{upper:.2f}')
plt.title('Log-vraisemblance en fonction de λ')
plt.xlabel('λ')
plt.ylabel('Log-vraisemblance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k à partir de la transformation logistique
x_k = np.log(p_k / (1 - p_k))

# Générer un échantillon à partir de la distribution logistique de scipy
x_logistic = stats.logistic.rvs(size=N)

# Créer des histogrammes pour les deux échantillons
bins = np.linspace(-5, 5, 50)
hist_k, _ = np.histogram(x_k, bins=bins, density=True)
hist_logistic, _ = np.histogram(x_logistic, bins=bins, density=True)

# Calculer les résidus (normalisés par l'incertitude)
# Uncertainty approximation
uncertainty_k = np.sqrt(hist_k * (N / np.sum(hist_k)))  # Scaling by total samples for normalization
uncertainty_logistic = np.sqrt(hist_logistic * (N / np.sum(hist_logistic)))

# Calculate residuals
residuals = (hist_k - hist_logistic) / (uncertainty_k + uncertainty_logistic)

# Tracer les histogrammes et les résidus
plt.figure(figsize=(14, 6))

# Histogrammes
plt.subplot(1, 2, 1)
plt.hist(x_k, bins=bins, density=True, alpha=0.5, color='blue', label='Échantillon personnalisé')
plt.hist(x_logistic, bins=bins, density=True, alpha=0.5, color='orange', label='Distribution logistique (SciPy)')
plt.title('Histogrammes des échantillons')
plt.xlabel('x')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)

# Histogramme des résidus
plt.subplot(1, 2, 2)
plt.bar(bins[:-1], residuals, width=np.diff(bins), alpha=0.7, color='purple', align='edge')
plt.axhline(0, color='red', linestyle='--')
plt.title('Histogramme des résidus normalisés')
plt.xlabel('Bins')
plt.ylabel('Résidus normalisés')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import minimize

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k à partir de la transformation logistique
x_k = np.log(p_k / (1 - p_k))

# Calculer la moyenne et la variance de {x_k}
mean_x = np.mean(x_k)
var_x = np.var(x_k)

# Construire le nouvel échantillon {z_k}
z_k = (x_k - mean_x) / np.sqrt(3 * var_x)

# Définir la fonction de log-vraisemblance négative pour λz
def negative_log_likelihood_z(λ, z_k):
    if λ <= 0:  # λ must be positive
        return np.inf
    likelihood = -np.sum(np.log((λ / 4) * (1 / np.cosh(λ * z_k / 2)**2)))
    return likelihood

# Estimer λz en minimisant la log-vraisemblance négative
initial_guess_z = 1.0
result_z = minimize(negative_log_likelihood_z, initial_guess_z, args=(z_k,), bounds=[(0.001, None)])
lambda_hat_z = result_z.x[0]

# Maximum log-likelihood
max_log_likelihood_z = -result_z.fun

# Évaluer la log-vraisemblance pour une plage de λ
lambda_values_z = np.linspace(0.1, 5, 100)
log_likelihoods_z = -np.array([negative_log_likelihood_z(λ, z_k) for λ in lambda_values_z])

# Trouver les valeurs de λ pour ∆lnL = 1/2
delta_lnL_z = max_log_likelihood_z - 0.5
lambda_range_z = lambda_values_z[(log_likelihoods_z >= delta_lnL_z)]

# Calculer l'incertitude σˆλz
if len(lambda_range_z) > 0:
    lambda_lower_z = lambda_range_z[0]
    lambda_upper_z = lambda_range_z[-1]
    sigma_lambda_hat_z = (lambda_upper_z - lambda_lower_z) / 2
else:
    sigma_lambda_hat_z = 0  # No valid range found

# Afficher les résultats
print(f"Estimation du paramètre λz : {lambda_hat_z}")
print(f"Incertitude estimée σˆλz : {sigma_lambda_hat_z}")

# Comparer avec la valeur connue de π
known_pi = np.pi
compatible = (lambda_hat_z - sigma_lambda_hat_z <= known_pi <= lambda_hat_z + sigma_lambda_hat_z)
print(f"Compatibilité avec π : {'Oui' if compatible else 'Non'}")

# Optionnel : tracer la log-vraisemblance
plt.figure(figsize=(10, 6))
plt.plot(lambda_values_z, log_likelihoods_z, label='Log-vraisemblance pour λz')
plt.axhline(delta_lnL_z, color='red', linestyle='--', label='ΔlnL = 1/2')
plt.axvline(lambda_hat_z, color='green', linestyle='--', label=f'λ̂z = {lambda_hat_z:.4f}')
plt.axvline(lambda_lower_z, color='orange', linestyle='--', label=f'λ_{lower:.4f}')
plt.axvline(lambda_upper_z, color='orange', linestyle='--', label=f'λ_{upper:.4f}')
plt.title('Log-vraisemblance en fonction de λz')
plt.xlabel('λz')
plt.ylabel('Log-vraisemblance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Nombre d'échantillons
N = 10**5

# Générer des probabilités uniformément distribuées
p_k = np.random.uniform(0, 1, N)

# Calculer x_k à partir de la transformation logistique
x_k = np.log(p_k / (1 - p_k))

# Définir la fonction de densité logistique
def logistic_pdf(x, λ):
    return (λ / 4) * (1 / np.cosh(λ * x / 2)**2)

# Paramètres
λ = 1
alpha_0 = 4 / np.sqrt(2 * np.pi)

# Calculer les densités
logistic_density = logistic_pdf(x_k, λ)
normal_density = norm.pdf(x_k, 0, 1)

# Calculer le ratio r(x_k)
r_k = normal_density / (alpha_0 * logistic_density)

# Générer un tableau pour les échantillons acceptés
accepted_samples = []

# Effectuer le processus d'acceptation
for i in range(N):
    p = np.random.uniform(0, 1)  # Générer un p uniformément distribué
    if r_k[i] >= p:  # Condition d'acceptation
        accepted_samples.append(x_k[i])

# Convertir en tableau NumPy
accepted_samples = np.array(accepted_samples)

# Ajuster une distribution normale
mu_hat, sigma_hat = norm.fit(accepted_samples)

# Tracer l'histogramme du sous-échantillon accepté
plt.figure(figsize=(10, 6))
plt.hist(accepted_samples, bins=50, density=True, alpha=0.5, color='blue', label='Histogramme des échantillons acceptés')

# Tracer la distribution normale ajustée
x_values = np.linspace(min(accepted_samples), max(accepted_samples), 1000)
pdf_fit = norm.pdf(x_values, mu_hat, sigma_hat)
plt.plot(x_values, pdf_fit, color='red', label=f'Normal fit: $\mu={mu_hat:.2f}$, $\sigma={sigma_hat:.2f}$')

# Ajouter des titres et des légendes
plt.title('Histogramme du sous-échantillon accepté avec ajustement normal')
plt.xlabel('x')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()

# Afficher les valeurs de fit
print(f"Valeur ajustée de µ : {mu_hat:.4f}")
print(f"Valeur ajustée de σ : {sigma_hat:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Nombre d'échantillons
N = 10**5

# Définir les paramètres de la distribution logistique
λ = 1

# Déterminer l'échelle optimale
# Le maximum de la densité logistique dans [a, b] est f(0; 1) = 1/4
alpha_0 = 4 / (10 - (-10))  # Échelle pour la plage [-10, 10]

# Générer des échantillons uniformes dans la plage [-10, 10]
uniform_samples = np.random.uniform(-10, 10, N)

# Calculer les densités logistique
def logistic_pdf(x, λ):
    return (λ / 4) * (1 / np.cosh(λ * x / 2)**2)

# Calculer les densités logistiques pour les échantillons uniformes
logistic_density = logistic_pdf(uniform_samples, λ)

# Calculer les densités de référence (uniforme)
uniform_density = np.ones_like(uniform_samples) * alpha_0

# Calculer le ratio r(x)
r = logistic_density / uniform_density

# Générer des échantillons acceptés
accepted_samples = []
for i in range(N):
    p = np.random.uniform(0, 1)  # Générer un p uniformément distribué
    if r[i] >= p:  # Condition d'acceptation
        accepted_samples.append(uniform_samples[i])

# Convertir en tableau NumPy
accepted_samples = np.array(accepted_samples)
M = len(accepted_samples)
acceptance_rate = M / N

# Afficher les résultats
print(f"Nombre total d'échantillons : {N}")
print(f"Nombre d'échantillons acceptés : {M}")
print(f"Taux d'acceptation M/N : {acceptance_rate:.4f}")

# Histogramme du sous-échantillon accepté
plt.figure(figsize=(10, 6))
plt.hist(accepted_samples, bins=50, density=True, alpha=0.5, color='blue', label='Histogramme des échantillons acceptés')

# Ajustement normal pour comparaison
mu_hat, sigma_hat = norm.fit(accepted_samples)
x_values = np.linspace(-10, 10, 1000)
pdf_fit = norm.pdf(x_values, mu_hat, sigma_hat)
plt.plot(x_values, pdf_fit, color='red', label=f'Ajustement Normal: $\mu={mu_hat:.2f}$, $\sigma={sigma_hat:.2f}$')

# Titres et légendes
plt.title('Histogramme des échantillons acceptés avec ajustement normal')
plt.xlabel('x')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()

# Optionnel: Utiliser une distribution normale tronquée comme référence
# Générer un échantillon à partir d'une distribution normale tronquée
truncated_normal_samples = np.random.normal(0, 1, N)
truncated_normal_samples = truncated_normal_samples[(truncated_normal_samples >= -10) & (truncated_normal_samples <= 10)]

# Répéter le processus d'acceptation pour la distribution normale tronquée
accepted_samples_truncated = []
for sample in truncated_normal_samples:
    p = np.random.uniform(0, 1)  # Générer un p uniformément distribué
    if logistic_pdf(sample, λ) / (alpha_0 * 0.1) >= p:  # Valeur de densité ajustée
        accepted_samples_truncated.append(sample)

# Convertir en tableau NumPy
accepted_samples_truncated = np.array(accepted_samples_truncated)
M_truncated = len(accepted_samples_truncated)
acceptance_rate_truncated = M_truncated / N

# Afficher les résultats de la distribution normale tronquée
print(f"Taux d'acceptation avec une distribution normale tronquée : M/N = {acceptance_rate_truncated:.4f}")
```
