```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.uniform(0, 1, size = 1000)
```

```python
plt.hist(data, bins = 100);
```

```python
plt.hist(data, bins = (0, 0.2, 0.5, 1));
```

```python
values, bins, art = plt.hist(data, bins=100)
```

```python
print(values)
print(bins)
print(art)
```

```python
data2 = np.random.normal(1, 2, size = 1000)
```

```python
fig = plt.figure()
bins = np.linspace(-10, 10, 200)
y, b, a = plt.hist(data, alpha=0.3, bins=bins, density=1, label='uniform')
y2, b2, a2 = plt.hist(data2, alpha=0.3, bins=bins, density=1, label='normal')
fig.legend();
plt.xlabel('X (cm)');
plt.ylabel('entries per bin');
fig.savefig('test.pdf')
```
