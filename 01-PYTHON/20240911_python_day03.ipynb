## SciPy

```python
from scipy import optimize
from scipy import stats
```

```python
x = np.random.uniform(-10, 10, size=6)
y = np.sin(x)
dy = abs(y)**0.5
```

```python
plt.scatter(x,y);
```

```python
plt.errorbar(x, y, yerr=dy, color="black", marker="s", linestyle="none");
```

```python
x_fine = np.linspace(-10, 10, 200)
y_fine = np.sin(x_fine)
```

```python
plt.errorbar(x, y, yerr=dy, color="black", marker="s", linestyle="none", label='data');
plt.plot(x_fine, y_fine, label='true function sin');
plt.legend()
```

```python
def poly1(x, a, b):
    y = a*x + b
    return y
```

```python
p, cov = optimize.curve_fit(poly1, x, y, sigma=dy)
```

```python
p
```

```python
poly_best = poly1(x_fine, p[0], p[1])
```

```python
plt.errorbar(x, y, yerr=dy, color="black", marker="s", linestyle="none", label='data');
plt.plot(x_fine, poly_best, label='fitted polynomial of order 1');
plt.legend()
```
